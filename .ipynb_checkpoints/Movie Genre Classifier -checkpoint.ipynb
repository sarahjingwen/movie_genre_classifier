# Movie Genre Classifier

This movie genre classifier aims to classify movies into their genre by looking at the movie synopsis.

## Import Libraries

```python
# For General Data Cleaning and Visualisation
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# For Text Cleaning
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import string

# For processing data into features
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from skmultilearn.problem_transform import ClassifierChain

# Metrics
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

## Load Data

```python
data = pd.read_csv("movies.csv")
data.head()
```

```python
data.dtypes
```

```python
data.info()
```

## Clean Genre Column

### Change Data Format of Genre to List

The current format of genre is a string, which is difficult be used for visualisation. Hence, I am going to change it into a list of genres.

```python
data["genre_list"] = data["genre"].map(lambda x: list(x.split(", ")))
data.head()
```

### List Unique Genres 

```python
unique_genres = list(set(genre for genre_list in data["genre_list"] for genre in genre_list))
unique_genres
```

```python
len(unique_genres)
```

There seems to be 21 different genres in the dataset.

### Barplot of Genre Frequencies

```python
freq = data['genre_list'].apply(lambda genre_list: pd.Series([genre for genre in genre_list])).stack().value_counts()
sns.barplot(y=freq.index, x=freq.values)
```

More than half the movies have the "Drama" genre tag. This would result in a class imbalance problem and may make the model biased. 

After doing some research in the "Drama" genre, it seems like almost any show could be considered a "Drama". Thus, I think that it is not very meaningful to have the "Drama" genre. Hence, I decided to take this genre out from the dataset.

### Removing "Drama" Genre

```python
# Checking to see if any movie only has drama as its genre
data[data.genre == "Drama"]
```

It seems like there would not be any movies having no genre if drama is removed.

```python
# Removing "Drama" Genre from genre list column
# Did not want to change existing genre column, so I did not remove drama from there

def remove_drama(genre_list):
    if "Drama" in genre_list:
        genre_list.remove("Drama")
    return genre_list.copy()

data["genre_list"] = data["genre_list"].apply(lambda genre_list: remove_drama(genre_list))
data.head()
```

```python
# Checking if Drama is successfullly removed by listing current unique genres
unique_genres = list(set(genre for genre_list in data["genre_list"] for genre in genre_list))
unique_genres
```

```python
# Plotting current genre frequency distribution
freq = data['genre_list'].apply(lambda genre_list: pd.Series([genre for genre in genre_list])).stack().value_counts()
sns.barplot(y=freq.index, x=freq.values)
```

The 3 most common genres are "Adventure", "Comedy" and "Action". There are still way more of these three genres than the rest, but would hope that this class imbalance would not be so serious.

## Clean Text

```python
def clean_text(text):
    # Convert to lower case
    text = text.lower()
    
    # Split by whitespace
    tokens = text.split()
    
    # Remove punctutation
    translation_table = str.maketrans('', '', string.punctuation)
    tokens = [token.translate(translation_table) for token in tokens]
    
    # Keep only alphabets
    tokens = [token for token in tokens if token.isalpha()]
    
    # Remove Stop words
    stop_words = stopwords.words('english')
    tokens = [token for token in tokens if not token in stop_words]
    
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    # Joining words back togather to pass into TF-IDF vectorizer
    text = ' '.join(tokens) 
    
    return text
```

```python
data["cleaned_synopsis"] = data["synopsis"].apply(lambda text: clean_text(text))
data.head()
```

## Convert to Features

### Convert Genre to Feature

```python
# Conduct one hot encoding for genre
multilabel_binarizer = MultiLabelBinarizer()
y = multilabel_binarizer.fit_transform(data['genre_list'])
y
```

```python
# Checking target variable y: seems like y is in the correct shape since it has 20 genres and 1431 rows
y.shape
```

### Convert Synopsis to Feature

Although there are better feature extraction methods out there, I would start simple first and use the Bag-Of-Words method.

```python
# Create count vectorizer
count_vectorizer = CountVectorizer()
```

```python
# Splitting that dataset into 80% training and 20% test set
x_train, x_test, y_train, y_test = train_test_split(data['cleaned_synopsis'], y, test_size=0.2, random_state=9)
```

```python
# Transforming the synopsis for training and test sets into Document Term Matrix
x_count_train = count_vectorizer.fit_transform(x_train)
x_count_test = count_vectorizer.transform(x_test)
```

## Training Models

Today, I am going to try out two models: OneVsRest and Chain Classifier. The OneVsRest classifier does not take into account the correlation between the genres, whereas the Chain Classifier takes into account the correlation between the genres. Based on the data, I highly suspect some genres are related. For example, adventure films tend to also be classified into the genres of action and fantasy.

The metrics I am going to be using is the F1 score, as well as precision and recall for each category. Given that this is a multilabel classification problem and the dataset is rather imbalanced, the f1 score would be a better measure than the accuracy score.

### OneVsRest with logistic regression

The OneVsRest strategy transforms the multi-lable problem into a multiple independent binary-class classification problem for each genre. 

```python
# Initialize 
classifier = OneVsRestClassifier(LogisticRegression())

# Fit
classifier.fit(x_count_train.toarray(), y_train)

# Predict
y_pred = classifier.predict(x_count_test)

# Check
print(y_pred[1])
print(multilabel_binarizer.inverse_transform(y_pred)[1])
```

```python
f1_score(y_test, y_pred, average="micro")
```

```python
print(classification_report(y_test, y_pred, zero_division=0))
```

At the default threshold of 0.5, the precision seems to be much higher than the recall. This means that the model does not want to take the risk of predicting wrongly and hence chooses not to predict when it does not know how to predict. 

The f1 score also seems to be quite low since it is closer to 0.

```python
# Finding the optimal threshold
y_pred_prob = classifier.predict_proba(x_count_test)
threshold = 0.2 # After much trial and error, this threshold seems good
y_pred_new = (y_pred_prob >= threshold).astype(int)
```

```python
f1_score(y_test, y_pred_new, average="micro")
```

```python
print(classification_report(y_test, y_pred_new,  zero_division=0))
```

The precision and recall seem to be more equal now and the f1 score has improved. 

### Classfier Chain

The classifier chain transforms the multi-label problem into a multiple independent binary-class classification problem. This is different from the previous method because the classifiers uses the predictions of the previous classifiers.

```python
# Initialize Classfier
chain_classifier = ClassifierChain(LogisticRegression())

# Fit Classifier
chain_classifier.fit(x_count_train, y_train)

# Predict
chain_y_pred = chain_classifier.predict(x_count_test)

# Check
print(chain_y_pred[1])
print(multilabel_binarizer.inverse_transform(chain_y_pred)[1])
```

```python
f1_score(y_test, chain_y_pred, average="micro")
```

```python
print(classification_report(y_test, chain_y_pred, zero_division=0))
```

Similar to the previous model, the precision is higher than the recall at the default threshold of 0.5. The f1 score is still rather low.

```python
# Finding the optimal threshold
chain_y_pred_prob = classifier.predict_proba(x_count_test)
threshold = 0.1 # After much trial and error, this threshold seems good
chain_y_pred_new = (chain_y_pred_prob >= threshold).astype(int)
```

```python
f1_score(y_test, chain_y_pred_new, average="micro")
```

```python
print(classification_report(y_test, chain_y_pred_new, zero_division=0))
```

To get a higher f1 score, the threshold value is 0.1. This makes the recall higher than the precision. This means that the model is more likely to give labels when it is not sure, resulting in less precise answers.

## Final Thoughts

The f1 scores for both models are rather similar. Hence, I cannot tell which model is better or whether there is really a correlation between the genres.

The f1 scores also seem rather low. This could perhaps be improved if I used better feature extraction methods for the synopsis, like word2vec (since the position of the words can play an important part in deciding the plot flow, which affects genre classification). I could also try out other models such as adapted algorithms like ml-knn or neural networks like LSTM.
